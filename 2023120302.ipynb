삼각형의 페르마 점(Fermat point)는 삼각형의 세 꼭짓점에서 거리의 합이 최소가 되는 점을 말한다. 이를 두 가지 시각에서 접근하려고 한다. 일단 작도를 활용할 수 있겠다. 다른 시각은 인공지능에 쓰이는 경사하강법이다. 이 두 가지를 수학공학도구인 Geogebra를 활용하여 보이려고 한다.

Geogebra를 실행시켜 경사하강법이다. 이 두 가지를 수학공학도구인 Geogebra를 활용하여 접근해 보겠다.

image.png

Geogebra를 실행시킨다.

image.png

[Font Size]를 [24pt]로 한다.

image.png

[Graphics]에서 마우스 오른쪽 버튼을 누르고 [Axes]를 선택하여 Axes를 보이지 않게 한다.

image.png

[Graphics]에서 마우스 오른쪽 버튼을 누르고 [Grid]를 선택하여 Grid를 보이지 않게 한다.

image.png

[Point] 도구를 선택한다.

image.png

[Graphics] 창에 마우스를 클릭하여 세 점을 만든다. 차례대로 $\rm{A}$, $\rm{B}$, $\rm{C}$의 이름이 붙여진다.

image.png

점 $\rm{A}$를 선택하고 오른쪽 마우스를 누르고, [Object Properties]를 선택한다.

image.png

[Properties] 창이 나타난다. 하지만, [Graphics] 창이 좁아진다. [Properties] 창의 닫기 버튼(x 모양) 근처에 가면 [Graphics] 창을 따로 띄울 수 있는 버튼이 나타난다. 이 버튼을 선택한다.

image.png

[Proferences] 창이 새로 나타났다. [Caption]에 \$\mathrm{A}\$ 를 입력한다. \$ 수식 \$ 형식으로 LaTex을 사용할 수 있다. [Show Label] 에 아래쪽 걱쇠를 선택하고 [Caption]을 선택한다. $\rm{A}$ 가 수학 책에서 보아왔던 점 이름 형태로 보인다.

image.png

점 $\rm{B}$, 점 $\rm{C}$를 마찮가지 방식으로 수학 책에서 보아왔던 형태로 바꿔준다.
